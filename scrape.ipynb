```python
url = 'https://www.historique-meteo.net/france/' # Le lien du site
print("Voici le lien du site contenant chaque region de france : "+ url)
```

```python
def get_all_region_and_relink(url: str) -> Tuple[List[str], List[str]]:
    """
    Brief:
    À partir d'une URL, récupère les noms de toutes les régions françaises et leur URL de redirection correspondante.
    
    Args:
    url: str, l'URL à partir de laquelle récupérer les noms de régions et les URL de redirection.
    
    Returns:
    Tuple[List[str], List[str]], une liste de noms de régions françaises et leur URL de redirection correspondante.
    """

    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    li_elements = soup.find_all('li', class_='item-thumbs')# Recupere les balise qui sont interessente fruit des recherhce du code html.

    region_names = []
    redirection_links = []
 
    for li in li_elements:
    ######## Avoir les Régions :
        region_name = li.find('div', class_='item-text').text.strip()
        region_names.append(region_name)
    ######## Avoir les liens de redirection : 
        redirection_link = li.find('a', class_='hover-wrap')['href']
        redirection_links.append(redirection_link)
    return (region_names, redirection_links)
```

```python
All_regions, relinks = get_all_region_and_relink(url=url)
nombre_de_ligne = len(All_regions)
print("Tout les noms de regions\t= ",All_regions, "\nrelink\t\t\t\t= ", relinks, "\nNombre de region\t\t=  " + str(nombre_de_ligne))
```

```python
def clear_relink(relinks:List[str]) -> None:
    """
    J'ajoute a chaque relink le prefix de l'url.
    """
    for i in range(len(relinks)):
        relinks[i] = url[0:-8] + relinks[i]

def clear_region_and_relink(All_regions:List[str], relinks:List[str], nombre_de_ligne:int) -> None :
    """
    Je verifier que dans All_regions, qu'il n'y as pas de \"\\n\" et on les remplace par des espace.\n
    Et\n
    J'ajoute a chaque relink le prefix de l'url.
    """
    clear_relink(relinks)
    for i in range(nombre_de_ligne):
        All_regions[i] = All_regions[i].replace("\n", " ")
def bref_affichage(All_regions:List[str], relinks:List[str], nombre_de_ligne:int, tour:int):
    """
    Affiche les \"tour\" premier tours l'id le nom d ela region et le lien
    """
    if (tour > len(All_regions)):
        print("la varaible tour est en overflow par raport aux données qu'on dispose.\nJe le rejuste aux maximum.")
        tour = len(All_regions)
    for test in range(11):
        print("ID = "+str(test)+", ", end='')
        if (test < 9):
            print(end=" ")
        print("Region = \""+All_regions[test]+"\",\tLien = "+relinks[test])
clear_region_and_relink(All_regions, relinks, nombre_de_ligne)
bref_affichage(All_regions, relinks, nombre_de_ligne, 15)
```

```python
# Recuperer toutes les autre information, tout les liens on a la meme structure.
# Ici on recuperer les année et les liens
def region_get_all_date_and_relink(lien_region: str) -> Tuple[List[str], List[str]]:
    """
    Pour une region de lien je vais voir quelle sont les date disponibles ensuite,
    je garderais les dates puis je donnerais les liens de chaque annés.
    """
    response = requests.get(lien_region)
    soup = BeautifulSoup(response.text, 'html.parser')

    per = soup.find('div', class_="list-group")# Recupere les balise qui sont interessente fruit des recherhce du code html.
    a_elements = per.find_all('a', class_='list-group-item')# Recupere les balise qui sont interessente fruit des recherhce du code html.

    années = []
    redirection_links = []
 
    for a in a_elements:
    ######## Avoir l'année :
        année = a.find('b').text.strip()
        années.append(année)
    ######## Avoir les liens de redirection : 
        redirection_link = a['href']
        redirection_links.append(redirection_link)
    clear_relink(redirection_links)
    return (années, redirection_links)
années, liens = region_get_all_date_and_relink(relinks[0])
print(années,"\n", liens)
def for_all_region_get_year_and_all_link(liens: List[str]) -> Tuple[List[List[str]], List[List[str]]]:
    """
    Pareil que region_get_all_date_and_relink mais pour toute les regions
    """
    ALL_Years = []
    ALL_LINK_YEAR = []
    for i in liens:
        year, link = region_get_all_date_and_relink(i)
        ALL_Years.append(year)
        ALL_LINK_YEAR.append(link)
    return (ALL_Years, ALL_LINK_YEAR)
Toute_les_années, tout_les_lien_dannée = for_all_region_get_year_and_all_link(relinks)
print(Toute_les_années,"\n", tout_les_lien_dannée)
```

```python
# print(tout_les_lien_dannée[0][0])
# On doit recupérer Les mois
def get_month_and_data(link: str) -> List[List[str]]:
    """
    Pour chaque moi retourne une liste contenant tout les informations importantes :\n
    Température moyenne | Température maximale | Température minimale | Température maximale maximum |
    Température minimale maximum | Température minimale minimum | Vitesse du vent | Température du vent |
    Précipitations moyennes par jour | Record de précipitations sur une journée | Humidité | Visibilité |
    Couverture nuageuse | Heure du lever du soleil | Heure du coucher du soleil | Durée du jour |
    """
    response = requests.get(link)
    soup = BeautifulSoup(response.text, 'html.parser')
    all_month_data = []
    pere = soup.find('div', id="contenu_page")
    all_data = pere.find_all("tbody")
    all_month = pere.find_all("h3")
    for i in range(len(all_month)):
        data = []
        data.append(all_month[i]["id"])
        #recuperer chaque data
        data_set = all_data[i]
        chiffre = data_set.find_all("b")
        data += ([chiffre_r.text for chiffre_r in chiffre])
        all_month_data.append(data)
    return (all_month_data)

# get_month_and_data(tout_les_lien_dannée[0][0])
# Maintenant qu'on as fais pour un lien nous devons le faire pour plusieur anné 
def for_all_year_get_month_and_data(links: List[str]) -> List[List[str]]:
    ALL_DATA_MONTH_BY_YEAR = []
    for l in links:
        ALL_DATA_MONTH_BY_YEAR.append(get_month_and_data(l))
    return(ALL_DATA_MONTH_BY_YEAR)
DATA_ETC_BY_YEAR = for_all_year_get_month_and_data((tout_les_lien_dannée[0]))
print(DATA_ETC_BY_YEAR)
# Maintenant qu'on as fais pour un lien dans un lien nous devons le faire pour les regions
def for_all_region_get_month_and_data(links: List[List[str]]) -> List[List[List[str]]]:
    ALL_DATA_MONTH_ETC_BY_REAGION = []
    for l in links:
        ALL_DATA_MONTH_ETC_BY_REAGION.append(for_all_year_get_month_and_data(l))
    return (ALL_DATA_MONTH_ETC_BY_REAGION)
# ALL_DATA_ETC = for_all_region_get_month_and_data(tout_les_lien_dannée) # Sa mets plus de 40min
```
